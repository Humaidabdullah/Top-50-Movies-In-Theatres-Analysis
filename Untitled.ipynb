# Adding metrics to DF


```python
df = pd.DataFrame(np.array(reviews),columns=["reviews"])
```

```python
#WordCount
df['word_count'] = df['reviews'].apply(lambda x: len(x.split()))
df.head()
```

```python
#Character count
df['char_count'] = df['reviews'].apply(lambda x : len(x))
df.head()
```

```python
#average length of each word inside the review
def average_words(x):
    words = x.split()
    return sum(len(word) for word in words) / len(words)
df['average_word_length'] = df['reviews'].apply(average_words)
df.head()
```

```python
#Stopwordcount
stop_words = stopwords.words('english')
```

```python
#Find how many stopwwords are there in each review
df['stopword_count'] = df['reviews'].apply(lambda x:len([word for word in x.split() if word.lower() in stop_words]))
df.head()
```

```python
#Stopword Rate
df['stopword_rate'] = df['stopword_count']/df['word_count']
df.head()
```

```python
df.describe()
```

# DATA CLEANING


```python
#Lowering the case
df['lowercase'] = df['reviews'].apply(lambda x: " ".join(word.lower() for word in x.split()))
df.head()
```

```python
#Removing punctuations
def rem_punc(review):
    review_blob = TextBlob(review)
    return ' '.join(review_blob.words)
df['punctuation'] = df['lowercase'].apply(rem_punc)
```

```python
#REMOVING STOPWORDS AND UNUSUAL SYMBOLS
def rem_stop_sym(review):
        review_list = [ele for ele in review.split()]
        clean_tokens = [t for t in review_list if re.match(r'[^\W\d)]*$',t)]
        clean_s = ' '.join(clean_tokens)
        clean_review = [word for word in clean_s.split() if word.lower() not in stopwords.words('english')]
        return " ".join(clean_review)
df['stopword'] = df['punctuation'].apply(rem_stop_sym)
```

```python
pd.Series(" ".join(df['stopword']).split()).value_counts()[:30]
```

```python
#Removing the words that do not contribute to the Sentiment Analysis
other_stopwords = ['also','like','get','would','one','even','way']
df['cleanreviews'] = df['stopword'].apply(lambda x: " ".join(word for word in x.split() if word not in other_stopwords))
df.head()
#Add other stopwords by repeating the pd.Series statement again for next 30 and so on
```

# Lemmatization

```python
df['lemmatized'] = df['cleanreviews'].apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
df.head()
```
